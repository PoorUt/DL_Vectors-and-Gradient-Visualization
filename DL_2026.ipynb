```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function and its partial derivatives
def f(x, y):
    return 4 - x**2 - 2*y**2

def partial_x(x, y):
    return -2 * x

def partial_y(x, y):
    return -4 * y

# Create meshgrid for plotting
x_vals = np.linspace(-2.5, 2.5, 100)
y_vals = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Two sample points and their gradient vectors
points = [(-1, 0.5), (1.5, -1)]
grads = [(partial_x(x, y), partial_y(x, y)) for x, y in points]

# Plot the surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, edgecolor='none')

# Add gradient vectors
for (x0, y0), (dx, dy) in zip(points, grads):
    z0 = f(x0, y0)
    ax.quiver(x0, y0, z0, dx, dy, 0, color='red', length=0.5, normalize=False)
    ax.scatter(x0, y0, z0, color='black')

# Set aspect ratio to make the surface appear natural
ax.set_box_aspect([5, 4, 2])  # x : y : z

# Labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D Plot of z = 4 - x² - 2y² with Gradient Vectors and Corrected Aspect Ratio')


```

```python
!pip install plotly numpy

import numpy as np
import plotly.graph_objects as go

# Define the function
def f(x, y):
    return 4 - x**2 - 2*y**2

# Create grid
x = np.linspace(-2.5, 2.5, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create surface
surface = go.Surface(x=X, y=Y, z=Z, colorscale='Viridis', opacity=0.8)

# Points for gradient vectors
points = [(-1, 0.5), (1.5, -1)]
arrows = []
lines = []

for x0, y0 in points:
    z0 = f(x0, y0)
    dx = -2 * x0
    dy = -4 * y0
    x1 = x0 + dx
    y1 = y0 + dy
    z1 = z0  # Gradient is in xy-plane only

    # Cone (arrowhead)
    arrows.append(
        go.Cone(
            x=[x0], y=[y0], z=[z0],
            u=[dx], v=[dy], w=[0],
            sizemode='absolute',
            sizeref=0.5,
            anchor='tail',
            showscale=False,
            colorscale='Reds'
        )
    )

    # Line (shaft of arrow)
    lines.append(
        go.Scatter3d(
            x=[x0, x1],
            y=[y0, y1],
            z=[z0, z1],
            mode='lines',
            line=dict(color='red', width=5),
            showlegend=False
        )
    )

# Layout
layout = go.Layout(
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z',
        aspectratio=dict(x=1.2, y=1, z=0.6)
    ),
    title='z = 4 - x² - 2y² with Gradient Vectors (Lines + Arrows)'
)

# Combine surface, cones, and lines
fig = go.Figure(data=[surface] + arrows + lines, layout=layout)
fig.show()
```

```python
!pip install plotly numpy

import numpy as np
import plotly.graph_objects as go

# Define function and partials
def f(x, y):
    return 4 - x**2 - 2*y**2

def grad(x, y):
    return (-2*x, -4*y)

# Grid points to animate through
x_vals = np.linspace(-2, 2, 10)
y_vals = np.linspace(-2, 2, 10)
points = [(x, y) for x in x_vals for y in y_vals]

# Surface
X, Y = np.meshgrid(np.linspace(-2.5, 2.5, 100), np.linspace(-2, 2, 100))
Z = f(X, Y)
surface = go.Surface(x=X, y=Y, z=Z, colorscale='Viridis', opacity=0.8)

# Build animation frames
frames = []
for (x0, y0) in points:
    z0 = f(x0, y0)
    dx, dy = grad(x0, y0)
    x1, y1, z1 = x0 + dx, y0 + dy, z0  # Gradient in XY plane

    frames.append(go.Frame(
        data=[
            surface,
            go.Cone(
                x=[x0], y=[y0], z=[z0],
                u=[dx], v=[dy], w=[0],
                sizemode="absolute", sizeref=0.5,
                anchor="tail", colorscale="Reds", showscale=False
            ),
            go.Scatter3d(
                x=[x0, x1], y=[y0, y1], z=[z0, z1],
                mode="lines", line=dict(color="red", width=5),
                showlegend=False
            )
        ],
        name=f"Point ({x0:.1f}, {y0:.1f})"
    ))

# Layout
layout = go.Layout(
    title="Animated Partial Derivatives of z = 4 - x² - 2y²",
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='z',
        aspectratio=dict(x=1.2, y=1, z=0.6)
    ),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(label='Play', method='animate', args=[None])]
    )]
)

# Build figure
fig = go.Figure(data=frames[0].data, frames=frames, layout=layout)
fig.show()
```
